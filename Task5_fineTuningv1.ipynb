#Create copy of datasets in colab for faster tuning

```python
import os, zipfile
from pathlib import Path

DRIVE_BASE = Path("/content/drive/MyDrive/FAIML_project_and_presentation/01_Project/large_files/datasets/cityscapes")
COLAB_DIR  = Path("/content/cityscapes")
LEFT_ZIP   = DRIVE_BASE / "leftImg8bit_trainvaltest.zip"
GT_ZIP     = DRIVE_BASE / "gtFine_trainvaltest.zip"

def cityscapes_ready(base):
    return (Path(base) / "leftImg8bit").is_dir() and (Path(base) / "gtFine").is_dir()

if cityscapes_ready(COLAB_DIR):
    print(f"✅ Dataset già pronto in {COLAB_DIR} — nessuna operazione.")
else:
    COLAB_DIR.mkdir(parents=True, exist_ok=True)
    for zip_path, label in [(LEFT_ZIP, "leftImg8bit"), (GT_ZIP, "gtFine")]:
        if not (COLAB_DIR / label).is_dir():
            print(f"📦 Estrazione {label}...")
            with zipfile.ZipFile(zip_path) as zf:
                zf.extractall(COLAB_DIR)
            print(f"   ✅ {label} estratto.")
        else:
            print(f"   ✅ {label} già presente, skip.")
    print("\n✅ Dataset pronto in", COLAB_DIR)
```
